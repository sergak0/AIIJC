```python
mitya_data = pd.read_csv('/content/drive/MyDrive/aiijc_sber/result_mitya.csv')
```

```python
mitya_desh = {0: "неизвестно",
              1: "животные",
              2: "музыка",
              3: "спорт",
              4: "литература"}
```

```python
mitya_data.ans = [mitya_desh[i] for i in mitya_data.ans]
```

```python
sergey_data = pd.read_csv('/content/drive/MyDrive/aiijc_sber/result_sergey.csv', delimiter=';')
```

```python
sergey_data.head()
```

```python
sergey_desh = {0: "неизвестно",
               1: "спорт",
               2: "литература",
               3: "животные",
               4: "музыка"}
```

```python
sergey_data.ans = [sergey_desh[i] for i in sergey_data.ans]
```

```python
sergey_data.head()
```

```python
all_texts = sergey_data.text.to_list() + lisa_data.text.to_list() + mitya_data.text.to_list()
all_ans = sergey_data.ans.to_list() + lisa_data.ans.to_list() + mitya_data.ans.to_list()
```

```python
total_ds = pd.DataFrame()
total_ds['text'] = all_texts
total_ds['ans'] = all_ans
```
